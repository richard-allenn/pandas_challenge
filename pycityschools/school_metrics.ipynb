```python
import pandas as pd

# File to Load (Remember to Change These)
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data)
student_data = pd.read_csv(student_data)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# count number of schools
number_of_schools = len(school_data)
```

```python
#count number of students
total_students = len(school_data_complete)
```

```python
#calculate total budget 
total_budget = sum(school_data["budget"])
```

```python
#calculate average maths score
average_maths_score = sum(student_data["maths_score"])/total_students
```

```python
# average reading score
average_reading_score = sum(student_data["reading_score"])/total_students
```

```python
#% of people passsing maths
passing_maths = sum(student_data["maths_score"] > 49.99)/total_students
```

```python
# % of people passing reading
passing_reading = sum(student_data["reading_score"] > 49.99)/total_students
```

```python
# % people passing overall
passing_overall = student_data[(student_data["reading_score"]> 49) &(student_data["maths_score"]> 49)].count()/total_students
```

```python
# Local goverbnment area summary
part_1 = {"total schools":number_of_schools,"total students":total_students,"total budget":total_budget, "average maths score":average_maths_score,"average reading score": average_reading_score,"% passing maths":passing_maths,"% passing reading": passing_reading,"% passing overall": passing_overall[2]}
LGA_summary = pd.DataFrame([part_1])
LGA_summary
```

```python
#total budget
school_data["per student budget"] = (
school_data["budget"] / school_data["size"])
```

```python
grouped_school = school_data_complete.groupby(school_data_complete["school_name"])
```

```python
#average reading score


reading_averagesdf = grouped_school["reading_score"].mean()

per_school1 = pd.merge(school_data, reading_averagesdf, on="school_name")
per_school1.head()
```

```python
print(grouped_school)
maths_averagesdf = grouped_school["maths_score",].mean()

per_school2 = pd.merge(per_school1, maths_averagesdf, on="school_name")
per_school2.head()
```

```python


stu_passingmath = student_data[student_data['maths_score'] > 49.99].groupby(['school_name'])['maths_score'].count()

per_school3 = pd.merge(per_school2, stu_passingmath, on="school_name")

per_school3['% passing math'] = (per_school3['maths_score_y'] /per_school3['size'])*100

school_summary = per_school3

school_summary
```

```python
#% passing raeding
stu_passingreading = student_data[student_data['reading_score'] > 49.99].groupby(['school_name'])['reading_score'].count()

per_school4 = pd.merge(per_school3, stu_passingreading, on="school_name")

per_school4['% passing reading'] = (per_school4['reading_score_y'] /per_school4['size'])*100

per_school4.head()
```

```python
per_school4.rename(columns={'type': 'school_type', 'size': 'total_students',"budget":"school_budget", 'reading_score_x':'average reading score','maths_score_x':'average maths score'}, inplace=True)

per_school5 = per_school4.drop(["maths_score_y","reading_score_y"], axis=1)

per_school5.head()
```
